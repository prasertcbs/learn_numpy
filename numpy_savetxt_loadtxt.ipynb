## savetxt (save ndarray to text file)
## loadtxt (load data from text file to ndarray)

```python
import numpy as np
print(f'numpy version = {np.__version__}')
```

```python
celsius=np.arange(0, 101)
fah=celsius * 1.8 + 32
fah
```

```python
celsius
```

```python
t=np.stack((celsius, fah), axis=1)
t
```

```python
np.savetxt('temperature.csv', t)
```

```python
np.savetxt('temperature2.csv', t, delimiter=',', fmt='%.1f')
```

```python
np.savetxt('temperature3.csv', t, 
           delimiter=',', 
           fmt='%.1f',
           header='celsius,fahrenheit')
```

```python
np.savetxt('temperature4.csv', t, 
           delimiter=',', 
           fmt='%.1f',
           header='celsius,fahrenheit',
           comments='')
```

## loadtxt

```python
s4=np.loadtxt('temperature4.csv', delimiter=',', skiprows=1)
s4
```

```python
s3=np.loadtxt('temperature3.csv', delimiter=',')
s3
```
