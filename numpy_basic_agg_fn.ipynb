```python
import numpy as np
print(f'numpy version = {np.__version__}')
```

```python
w=np.array([70, 48, 52, 60, 52])
w
```

```python
w.size
```

```python
np.min(w)
```

```python
np.max(w)
```

```python
np.sum(w)
```

```python
np.mean(w)
```

```python
np.median(w)
```

```python
np.sort(w)
```

## 2-dim array

```python
a=np.reshape(np.arange(1, 21), (4, 5))
a
```

```python
np.mean(a)
```

```python
a.size
```

```python
a.shape
```

```python
np.mean(a[0])
```

```python
np.mean(a[:, 1])
```

```python
a
```

```python
np.mean(a, axis=0)
```

```python
np.sum(a, axis=0)
```

```python
np.sum(a, axis=1)
```

```python
np.std(a)
```

```python
np.var(a)
```

```python
np.ptp(a) # peak to peak
```

```python
np.max(a) - np.min(a)
```
