# reshape, ravel, tile, repeat

```python
import numpy as np
print(f'numpy version = {np.__version__}')
```

## reshape

```python
a=np.arange(1, 13)
a
```

```python
a.reshape(4, 3)
```

```python
a.reshape(4, -1)
```

```python
a.reshape(-1, 3)
```

```python
a.reshape(-1, 2)
```

```python
b=a.reshape(-1, 2)
b
```

```python
b.reshape(1, -1)
```

```python
b.ravel()
```

## tile

```python
d=np.array([2, 1, 8])
d
```

```python
np.tile(d, 3)
```

```python
np.tile(d, (3, 1))
```

```python
e=np.array([[1, 2], [10, 20]])
e
```

```python
np.tile(e, 3)
```

```python
np.tile(e, (5, 1))
```

```python
np.tile(e, (5, 2))
```

```python
np.tile([[4, 7], [8, 9]], (4, 3))
```

```python
np.tile([1, 0], 4)
```

```python
np.tile([[1, 0], [0, 1]], (4, 4))
```

## repeat

```python
np.repeat(9, 3)
```

```python
np.repeat([10, 15, 20], 2)
```

```python
g=np.array([[1,2], [8, 9]])
g
```

```python
np.repeat(g, 3)
```

```python
g.ravel()
```

```python
np.repeat(g.ravel(), 3)
```
