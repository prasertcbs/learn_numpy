# Iterate NumPy array

```python
import numpy as np
print(f'NumPy version = {np.__version__}')
```

```python
a=np.arange(10) * 10
a
```

```python
for n in a:
    print(n)
```

```python
for i, v in enumerate(a):
    print(i, v)
```

```python
for i, v in enumerate(a, start=1):
    print(i, v)
```

## 2-dim array

```python
b=np.arange(1, 21).reshape(4, 5)
b
```

```python
for r in b:
    print(r)
```

```python
for n in b.flat:
    print(n)
```

```python
b
```

```python
for i, v in np.ndenumerate(b):
    print(i, v)
```

```python
b
```

```python
for i, v in np.ndenumerate(b[1]):
    print(i, v)
```

```python
for i, v in np.ndenumerate(b[1:4]):
    print(i, v)
```

```python
for i, v in np.ndenumerate(b[:, 2]):
    print(i, v)
```

```python
b
```

```python
for n in b.flat:
    print(n)
```
